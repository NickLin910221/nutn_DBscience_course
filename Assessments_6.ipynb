```
install.packages("SportsAnalytics")
install.packages("data.table")
install.packages("ggplot2")
library(SportsAnalytics)
library(data.table)
library(ggplot2)
```

```
NBA1819 = fetch_NBAPlayerStatistics("18-19")
```

```
NBA1819_PG = subset(NBA1819,Position = "PG")
str(NBA1819_PG)
```

```
### True : 三分球得分數>200,很強的！False : 就普通人而已啦！
qplot(ThreesAttempted,ThreesMade,data = NBA1819_PG,color = ThreesMade>200)
qplot(Team,ThreesMade,data = NBA1819_PG,color = ThreesMade>200)
qplot(FreeThrowsMade,FreeThrowsAttempted,data = NBA1819_PG,color = ThreesMade>200)
qplot(Team,FreeThrowsAttempted,data = NBA1819_PG,color = ThreesMade>200)
```

```
qplot(ThreesAttempted,data = NBA1819_PG,fill = Team,binwidth = 100)
qplot(ThreesMade,data = NBA1819_PG,fill = Team,binwidth = 50)
qplot(FreeThrowsMade,data = NBA1819_PG,fill = Team,binwidth = 100)
qplot(FreeThrowsAttempted,data = NBA1819_PG,fill = Team,binwidth = 50)
```
