```
install.packages("SportsAnalytics")
library(SportsAnalytics)
install.packages("ggplot2")
library(ggplot2)
```

```
### 抓取NBA 19-20年度球員資料
NBA1920 = fetch_NBAPlayerStatistics("19-20")
str(NBA1920)
```

```
### 利用order()去抓取19-20賽季所有球員三分球得分數前30高
NBA1920_ThreesMade = NBA1920[order(NBA1920$ThreesMade,decreasing = T),]
NBA1920_ThreesMade[1:30,]
NBA1920_FieldGoalsMade = NBA1920[order(NBA1920$FieldGoalsMade,decreasing = T),]
NBA1920_FieldGoalsMade[1:30,]
```

```
### 利用subset()過濾出隊伍為達拉斯小牛隊的球員
subset(NBA1920,Team=="DAL")
```

```
### 找出我最愛的球員 Luka Doncic
NBA1920[grepl("Luka Doncic",NBA1920$Name),]
### 這邊為意外的怪事?(註一)
NBA1920[grepl("CLI",NBA1920$Team),]
NBA1920[grepl("NA",NBA1920$Team),]
```

```
### 計算各隊球員人數，但其中有兩隊人數非常怪異(註一)
ggplot(NBA1920,aes(x=Team)) + geom_bar()
### 把各隊球員三分球得分做成一個點，不意外的那個接近300分的點就是HOU的James Harden
ggplot(NBA1920,aes(x=Team,y=ThreesMade)) + geom_point()
### 把球員三分球數量做成箱型圖可發現大部分球員三分球進球數大多位於0-75居多
ggplot(NBA1920,aes(ThreesMade)) + geom_boxplot()
### 將X軸設為三分球出手次數，Y軸設為三分球進球數，我們可以看到命中率其實是一條斜直線但進球數接近300分那點代表著，大鬍子會進那麼多球只是HOU的進攻策略為拿到球就給大鬍子投三分啦
ggplot(NBA1920,aes(x=ThreesAttempted,y=ThreesMade)) + geom_count()
```
